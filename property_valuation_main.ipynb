**Project Folder Structure**

```python
import os

BASE_DIR = "/content/drive/MyDrive/property_valuation"

folders = [
    "data",
    "images",
    "outputs",
    "notebooks",
    "scripts"
]

for f in folders:
    os.makedirs(os.path.join(BASE_DIR, f), exist_ok=True)

BASE_DIR
```

**Load the Dataset**

```python
import pandas as pd

DATA_PATH = f"/content/drive/MyDrive/property_valuation/kc_house_data.csv"
df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'lat', 'long']].head()
```

**Split the Dataset**

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42
)
```

```python
train_df.to_csv(f"{BASE_DIR}/data/train.csv", index=False)
test_df.to_csv(f"{BASE_DIR}/data/test.csv", index=False)
```
